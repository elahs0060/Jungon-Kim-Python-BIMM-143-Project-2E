# Part 2: Load in the data and perform Bioinformatics Analyses (Pairwise sequence alignment & Protein structure analysis )

Fasta file is a text based data file that contains a single line description with single letter sequence. For this project, I downloaded all the T1R1 FASTA files from Panda, Human, Mouse, Rat, Dog, and Catsaved from Uniport database for pairwise sequence alignment. 

In the code below, we read in protein seuqence from the FASTA file of Panda and develope global pairwise sequence alignment with other organisms such as Human, Mouse, Rat, Dog, and Cat. After pairwise sequence alignment, I print the length of the alignment to check the size and print the pairwise sequence alignment with alignment scores. At the end, we record the match scores from the alignments and print them for clarification. Then, I merge all the Fasta files into one big file for the Heatmap clustering.

```python
#Pairwise Sequence alignment

#Panda sequence
#Read in T1R1 Panda sequence from the Fasta file saved from Uniport for alignment
FastaPanda = SeqIO.parse('Panda.fasta.txt', 'fasta')
FastaPandaR = SeqIO.read('Panda.fasta.txt', 'fasta')

#Human sequence
#Read in T1R1 Human sequence from the Fasta file saved from Uniport for alignment
FastaHuman = SeqIO.parse('Human.fasta.txt', 'fasta')
FastaHumanR = SeqIO.read('Human.fasta.txt', 'fasta')

#Mouse sequence
#Read in T1R1 Mouse sequence from the Fasta file saved from Uniport for alignment
FastaMouse = SeqIO.parse('Mouse.fasta.txt', 'fasta')
FastaMouseR = SeqIO.read('Mouse.fasta.txt', 'fasta')

#Rat sequence
#Read in T1R1 Rat sequence from the Fasta file saved from Uniport for alignment
FastaRat = SeqIO.parse('Rat.fasta.txt', 'fasta')
FastaRatR = SeqIO.read('Rat.fasta.txt', 'fasta')

#Dog sequence
#Read in T1R1 Dog sequence from the Fasta file saved from Uniport for alignment
FastaDog = SeqIO.parse('Dog.fasta.txt', 'fasta')
FastaDogR = SeqIO.read('Dog.fasta.txt', 'fasta')

#Cat sequence
#Read in T1R1 Cat sequence from the Fasta file saved from Uniport for alignment
FastaCat = SeqIO.parse('Cat.fasta.txt', 'fasta')
FastaCatR = SeqIO.read('Cat.fasta.txt', 'fasta')

#Save a global pairwise sequence between the Panda and other Organisms
PandaHumanalignment = pairwise2.align.globalxx(FastaPandaR.seq, FastaHumanR.seq)
PandaMousealignment = pairwise2.align.globalxx(FastaPandaR.seq, FastaMouseR.seq)
PandaRatalignment = pairwise2.align.globalxx(FastaPandaR.seq, FastaRatR.seq)
PandaDogalignment = pairwise2.align.globalxx(FastaPandaR.seq, FastaDogR.seq)
PandaCatalignment = pairwise2.align.globalxx(FastaPandaR.seq, FastaCatR.seq)

#print the lengths of the alifnment of Panda to other organisms
print(len(PandaHumanalignment))
print(len(PandaMousealignment))
print(len(PandaRatalignment))
print(len(PandaDogalignment))
print(len(PandaCatalignment))

#print pairwise sequence alignment of Panda to other organisms
for a in PandaHumanalignment:
    print(format_alignment(*a))
for a in PandaMousealignment:
    print(format_alignment(*a))
for a in PandaRatalignment:
    print(format_alignment(*a))
for a in PandaDogalignment:
    print(format_alignment(*a))
for a in PandaCatalignment:
    print(format_alignment(*a))

#Make a bracket for stating match scores of Pirwise sequence alignment
match_scores = []
aligner = Align.PairwiseAligner()    

#Record a score for Pairwise sequence alignment of Panda to other organisms
score = aligner.score(FastaPandaR.seq, FastaHumanR.seq)
match_scores.append(score)
score = aligner.score(FastaPandaR.seq, FastaMouseR.seq)
match_scores.append(score)
score = aligner.score(FastaPandaR.seq, FastaRatR.seq)
match_scores.append(score)
score = aligner.score(FastaPandaR.seq, FastaDogR.seq)
match_scores.append(score)
score = aligner.score(FastaPandaR.seq, FastaCatR.seq)
match_scores.append(score)

#Print score sfor Pairwise sequence alignment of Panda to other organisms
print(match_scores)
```

```python
#Merge All Fasta file into one file for Heatmap Clustering
filenames = ['Panda.fasta.txt', 'Human.fasta.txt', 'Mouse.fasta.txt', 'Rat.fasta.txt', 'Dog.fasta.txt','Cat.fasta.txt',]
with open('All_File.fasta.txt', 'w') as outfile:
    for fname in filenames:
        with open(fname) as infile:
            outfile.write(infile.read())
```

# Part 3: Create a heatmap (plotting method) to show the Pairwise sequence alignment scores (bioinformatics method) between all T1R1 genes. 

A heatmap is a bio informatic plotting method that uses the system of color coding to represent different values of two inputs. In this project, the inputs are Panda T1R1 gene sequence and T1R1 Human, Mouse, Rat, Dog, or Cat gene seuqence. The colors represent the alignment score between the sequences in the inputs.

```python
#Heatplot for pairwise sequence alignment
input_file = 'All_File.fasta.txt'
Fasta = list(SeqIO.parse("All_File.fasta.txt", "fasta"))

size = len(Fasta)
#Empty numpy matrix is filled with the length of the merged fasta file
alignments_score = np.empty(shape=(size,size))
```

```python
#Nesterd for loop that functions as pairwise sequence alignment  between Panda gene and other organisms
def pairwise_seq_heatmap(my_numpy_array, size):
    for i in range(size):
        for j in range(size):
            X = Fasta[i].seq
            Y = Fasta[j].seq
            print(Fasta[i].id)
        
         # Lisst of the global alignments between the sequences Panda and other organisms
            my_numpy_array[i,j] = int(pairwise2.align.globalxx(X, Y, score_only=True))
    return my_numpy_array
#Checking the creation of the matrix through scores
My_alignment_Scores = pairwise_seq_heatmap(alignments_score, size)
```

For this project, I use the input file called All_File. It is the merged Fasta file of the Fasta files that has been used in pairwise sequence alignment. For the plots, we use matplotlib. The X and Y are the sequences that should be aligned to. The subplot is created and heatplot is now displayed. Heatmap also includes the axes and colorbar is also added, so viewers can understand the alignment score based on the colors. For this project, heatmap is used to represent the pairwise sequence alignement score between Panda's T1R1 and different organisms' T1R1. It is used to find out the difference of Panda's T1R1 to other organisms'T1R1 to understand its uniquness.

```python
#Creation of subplot
fig, ax = plt.subplots()

# Display of heatplot
plt.imshow(alignments_score)

# Creating the heat map to show axes
im = ax.imshow(alignments_score)

# Addition of colorbar next to the heatmap
cbar = ax.figure.colorbar(im)

# Display of the plot
plt.show()
```
